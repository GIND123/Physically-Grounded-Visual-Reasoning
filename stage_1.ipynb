```python
import os

def print_tree(root, prefix=""):
    try:
        items = sorted(os.listdir(root))
    except PermissionError:
        print(prefix + "└── [Permission Denied]")
        return

    n = len(items)

    for i, name in enumerate(items):
        path = os.path.join(root, name)
        connector = "└── " if i == n - 1 else "├── "

        # If directory has many children → compress
        if os.path.isdir(path):
            try:
                sub_items = sorted(os.listdir(path))
            except:
                sub_items = []

            if len(sub_items) > 2:
                print(prefix + connector + name + "/")
                
                # Print first
                print(prefix + "    ├── " + sub_items[0])
                # Ellipsis
                print(prefix + "    ├── ...")
                # Print last
                print(prefix + "    └── " + sub_items[-1])
            else:
                print(prefix + connector + name + "/")
                print_tree(path, prefix + ("    " if i == n - 1 else "│   "))
        else:
            print(prefix + connector + name)


# Run on your dataset
print_tree("Physically-Grounded-Visual-Reasoning")
```

```python
# =============================================================================
# PGVR-IDA STAGE 1: MODEL TRAINING & FOUNDATION
# =============================================================================
# Run in Google Colab with A100 GPU (or T4 with reduced batch size)
# Mount Google Drive first: drive.mount('/content/drive')
#
# This notebook:
#   Cell 1:  Setup & installs
#   Cell 2:  Build training triplets from MVTec AD
#   Cell 3:  Fine-tune SD Inpainting LoRA (per-category)
#   Cell 4:  Validate trained LoRA models (FID, LPIPS, SSIM, visual)
#   Cell 5:  ControlNet edge-conditioned inpainting setup
#   Cell 6:  SAM precise mask generation module
#   Cell 7:  Proper PatchCore & EfficientAD baselines via anomalib
#   Cell 8:  Extract & index all DINOv2 features (train + test)
#   Cell 9:  Stage 1 summary & checkpoint inventory
# =============================================================================


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 1: SETUP & INSTALLS                                               ║
# ╚═══════════════════════════════════════════════════════════════════════════╝

# !pip install -q diffusers[torch] accelerate transformers peft bitsandbytes
# !pip install -q torchmetrics lpips clean-fid pillow scikit-learn scipy
# !pip install -q open-clip-torch segment-anything opencv-python-headless
# !pip install -q anomalib faiss-gpu tqdm wandb
# !pip install -q controlnet-aux

import os, json, gc, sys, shutil, random, time, math, warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from pathlib import Path
from PIL import Image
from tqdm.auto import tqdm
from collections import defaultdict
from datetime import datetime

warnings.filterwarnings("ignore")

# ── Paths ────────────────────────────────────────────────────────────────────
ROOT         = "/root/project"
MVTEC        = f"{ROOT}/datasets/mvtec_ad"
CFG_DIR      = f"{ROOT}/configs"
CKPT_DIR     = f"{ROOT}/checkpoints"
SYN_DIR      = f"{ROOT}/synthetic_outputs"
RES_DIR      = f"{ROOT}/results"
KB_DIR       = f"{ROOT}/knowledge_base"
HYPO_DIR     = f"{ROOT}/hypotheses"
FEATURES_DIR = f"{CKPT_DIR}/features"

# New Stage 1 directories
TRIPLET_DIR  = f"{ROOT}/training_triplets"
LORA_DIR     = f"{CKPT_DIR}/sd_lora"
CNET_DIR     = f"{CKPT_DIR}/controlnet"
DETECTOR_DIR = f"{CKPT_DIR}/detectors"
EVAL_S1_DIR  = f"{RES_DIR}/stage1_validation"

for d in [TRIPLET_DIR, LORA_DIR, CNET_DIR, DETECTOR_DIR, EVAL_S1_DIR]:
    os.makedirs(d, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {DEVICE}")
if DEVICE.type == "cuda":
    print(f"GPU: {torch.cuda.get_device_name()}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")

# Load taxonomy
with open(f"{CFG_DIR}/defect_taxonomy.json") as f:
    TAXONOMY = json.load(f)
ALL_CATEGORIES = list(TAXONOMY["mvtec_ad"].keys())
print(f"\nCategories ({len(ALL_CATEGORIES)}): {ALL_CATEGORIES}")

IMG_SIZE = 512


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 2: BUILD TRAINING TRIPLETS FROM MVTec AD                          ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# For each (category, defect_type), create triplets:
#   (nearest_normal_image, ground_truth_mask, defective_image)
# The "nearest normal" is found by DINOv2 CLS feature similarity.
# ═══════════════════════════════════════════════════════════════════════════════

print("=" * 80)
print("CELL 2: Building training triplets")
print("=" * 80)

from torchvision import transforms

dinov2_transform = transforms.Compose([
    transforms.Resize((518, 518)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Load DINOv2 for feature matching
print("Loading DINOv2 ...")
dinov2 = torch.hub.load("facebookresearch/dinov2", "dinov2_vitb14", verbose=False)
dinov2 = dinov2.eval().to(DEVICE)
print("✓ DINOv2 loaded")


@torch.no_grad()
def extract_cls_feature(img_path):
    """Extract DINOv2 CLS token for a single image."""
    img = Image.open(img_path).convert("RGB")
    x = dinov2_transform(img).unsqueeze(0).to(DEVICE)
    out = dinov2.forward_features(x)
    return out["x_norm_clstoken"].squeeze(0).cpu()


def get_image_paths(category, split, subset):
    img_dir = os.path.join(MVTEC, category, split, subset)
    if not os.path.exists(img_dir):
        return []
    exts = {".png", ".jpg", ".jpeg", ".bmp", ".tiff"}
    return sorted([
        os.path.join(img_dir, f) for f in os.listdir(img_dir)
        if os.path.splitext(f)[1].lower() in exts
    ])


def find_nearest_normal(defect_feat, normal_feats, normal_paths):
    """Find the normal image most similar to the defect image."""
    sims = F.cosine_similarity(defect_feat.unsqueeze(0), normal_feats, dim=1)
    best_idx = sims.argmax().item()
    return normal_paths[best_idx], sims[best_idx].item()


triplet_stats = {}

for cat in ALL_CATEGORIES:
    cat_triplet_dir = f"{TRIPLET_DIR}/{cat}"
    os.makedirs(cat_triplet_dir, exist_ok=True)

    # Check if already built
    manifest_path = f"{cat_triplet_dir}/manifest.json"
    if os.path.exists(manifest_path):
        with open(manifest_path) as f:
            existing = json.load(f)
        print(f"  {cat}: {len(existing['triplets'])} triplets (cached)")
        triplet_stats[cat] = len(existing["triplets"])
        continue

    print(f"\n── {cat} ──")

    # Get normal training images and extract features
    normal_paths = get_image_paths(cat, "train", "good")
    if not normal_paths:
        print(f"  ⚠ No normal training images, skipping")
        continue

    # Try to load cached CLS features
    cached_cls = f"{FEATURES_DIR}/{cat}/cls_features.pt"
    if os.path.exists(cached_cls):
        normal_feats = torch.load(cached_cls, map_location="cpu")
        # Ensure shape matches
        if normal_feats.shape[0] != len(normal_paths):
            print(f"  Cached features shape mismatch, re-extracting...")
            normal_feats = torch.stack([extract_cls_feature(p) for p in tqdm(normal_paths, desc="  Normal feats")])
    else:
        print(f"  Extracting CLS features for {len(normal_paths)} normal images...")
        normal_feats = torch.stack([extract_cls_feature(p) for p in tqdm(normal_paths, desc="  Normal feats")])

    # Find all defect types
    test_dir = os.path.join(MVTEC, cat, "test")
    gt_dir = os.path.join(MVTEC, cat, "ground_truth")
    defect_types = [
        d for d in sorted(os.listdir(test_dir))
        if d != "good" and os.path.isdir(os.path.join(test_dir, d))
    ]

    triplets = []

    for dt in defect_types:
        defect_paths = get_image_paths(cat, "test", dt)
        mask_dir = os.path.join(gt_dir, dt)

        for dp in defect_paths:
            fname = os.path.splitext(os.path.basename(dp))[0]
            mask_path = os.path.join(mask_dir, f"{fname}_mask.png")

            if not os.path.exists(mask_path):
                continue

            # Validate mask has actual defect pixels
            mask = np.array(Image.open(mask_path).convert("L"))
            defect_ratio = (mask > 127).sum() / mask.size
            if defect_ratio < 0.001 or defect_ratio > 0.7:
                continue  # Skip degenerate masks

            # Find nearest normal
            defect_feat = extract_cls_feature(dp)
            nearest_normal, sim = find_nearest_normal(defect_feat, normal_feats, normal_paths)

            triplets.append({
                "defect_image": dp,
                "mask_image": mask_path,
                "normal_image": nearest_normal,
                "category": cat,
                "defect_type": dt,
                "similarity": round(sim, 4),
                "defect_ratio": round(defect_ratio, 4),
            })

    # Save manifest
    manifest = {
        "category": cat,
        "n_triplets": len(triplets),
        "n_defect_types": len(defect_types),
        "defect_types": defect_types,
        "created": datetime.now().isoformat(),
        "triplets": triplets,
    }
    with open(manifest_path, "w") as f:
        json.dump(manifest, f, indent=2)

    triplet_stats[cat] = len(triplets)
    print(f"  ✓ {len(triplets)} triplets across {len(defect_types)} defect types")

# Summary
print(f"\n{'═' * 60}")
print("TRIPLET SUMMARY")
print(f"{'═' * 60}")
total = 0
for cat in ALL_CATEGORIES:
    n = triplet_stats.get(cat, 0)
    total += n
    print(f"  {cat:<15} {n:>4} triplets")
print(f"{'─' * 60}")
print(f"  {'TOTAL':<15} {total:>4} triplets")

# Unload DINOv2 to free VRAM for SD training
del dinov2
gc.collect()
torch.cuda.empty_cache()
print("\n✓ DINOv2 unloaded, VRAM freed for Stage 1B")


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 3: FINE-TUNE SD INPAINTING LoRA (PER-CATEGORY)                    ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# Trains a LoRA adapter on top of SD 2 Inpainting for each MVTec category.
# Each adapter learns the visual vocabulary of that category's real defects.
#
# Input:  training triplets from Cell 2
# Output: checkpoints/sd_lora/{category}/  (LoRA weights + config)
# ═══════════════════════════════════════════════════════════════════════════════

print("=" * 80)
print("CELL 3: Fine-tuning SD Inpainting LoRA")
print("=" * 80)

from diffusers import StableDiffusionInpaintPipeline, DDPMScheduler
from diffusers import AutoencoderKL, UNet2DConditionModel
from transformers import CLIPTextModel, CLIPTokenizer
from peft import LoraConfig, get_peft_model, PeftModel
from torch.utils.data import Dataset, DataLoader

# ── Configuration ────────────────────────────────────────────────────────────
LORA_CONFIG = {
    "rank": 8,
    "alpha": 16,
    "dropout": 0.05,
    "target_modules": [
        "to_q", "to_k", "to_v", "to_out.0",  # attention
        "proj_in", "proj_out",                  # projection
    ],
    "learning_rate": 1e-4,
    "weight_decay": 1e-2,
    "max_steps_per_category": 1500,
    "min_steps_per_category": 400,
    "steps_per_sample": 15,       # scale steps with dataset size
    "batch_size": 1,
    "grad_accum": 4,
    "warmup_ratio": 0.1,
    "save_every": 500,
    "resolution": 512,
    "mixed_precision": True,
}

# Categories to train (set to ALL_CATEGORIES for full run, or subset for testing)
TRAIN_CATEGORIES = ALL_CATEGORIES  # or e.g. ["bottle", "metal_nut", "transistor"]


# ── Dataset ──────────────────────────────────────────────────────────────────

class InpaintingTripletDataset(Dataset):
    """
    Loads (normal, mask, defect) triplets for SD inpainting training.
    The model learns: given normal + mask → produce defect in masked region.
    """
    def __init__(self, manifest_path, resolution=512):
        with open(manifest_path) as f:
            data = json.load(f)
        self.triplets = data["triplets"]
        self.category = data["category"]
        self.resolution = resolution

        self.img_transform = transforms.Compose([
            transforms.Resize((resolution, resolution), interpolation=transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize([0.5], [0.5]),  # SD expects [-1, 1]
        ])
        self.mask_transform = transforms.Compose([
            transforms.Resize((resolution, resolution), interpolation=transforms.InterpolationMode.NEAREST),
            transforms.ToTensor(),
        ])

    def __len__(self):
        return len(self.triplets)

    def __getitem__(self, idx):
        t = self.triplets[idx]

        # Load images
        normal = Image.open(t["normal_image"]).convert("RGB")
        defect = Image.open(t["defect_image"]).convert("RGB")
        mask = Image.open(t["mask_image"]).convert("L")

        # Transform
        normal_tensor = self.img_transform(normal)    # [-1, 1], 3×H×W
        defect_tensor = self.img_transform(defect)    # [-1, 1], 3×H×W
        mask_tensor = self.mask_transform(mask)        # [0, 1], 1×H×W

        # Binarize mask (>0.5 = defect region)
        mask_tensor = (mask_tensor > 0.5).float()

        # Create masked normal (set defect region to 0 — this is what SD inpainting expects)
        masked_normal = normal_tensor * (1 - mask_tensor)

        # Build text prompt
        defect_type = t["defect_type"].replace("_", " ")
        category = t["category"].replace("_", " ")
        prompt = (
            f"a {defect_type} defect on a {category}, "
            f"industrial inspection photograph, macro, high resolution"
        )

        return {
            "normal": normal_tensor,
            "defect": defect_tensor,           # this is the TARGET
            "mask": mask_tensor,
            "masked_normal": masked_normal,
            "prompt": prompt,
            "defect_type": t["defect_type"],
        }


# ── Training Loop ────────────────────────────────────────────────────────────

def train_lora_for_category(category, config):
    """Train a LoRA adapter for one MVTec category."""

    cat_lora_dir = f"{LORA_DIR}/{category}"
    done_flag = f"{cat_lora_dir}/.done"

    # Skip if already trained
    if os.path.exists(done_flag):
        print(f"  {category}: LoRA already trained (skipping)")
        return True

    manifest_path = f"{TRIPLET_DIR}/{category}/manifest.json"
    if not os.path.exists(manifest_path):
        print(f"  {category}: No triplets found (skipping)")
        return False

    with open(manifest_path) as f:
        manifest = json.load(f)
    n_triplets = manifest["n_triplets"]
    if n_triplets < 5:
        print(f"  {category}: Only {n_triplets} triplets — too few (skipping)")
        return False

    os.makedirs(cat_lora_dir, exist_ok=True)

    # Calculate training steps
    steps = min(
        config["max_steps_per_category"],
        max(config["min_steps_per_category"], n_triplets * config["steps_per_sample"])
    )

    print(f"\n  {category}: {n_triplets} triplets → {steps} steps")

    # ── Load base model ──────────────────────────────────────────────────────
    model_id = "sd2-community/stable-diffusion-2-inpainting"

    tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer")
    text_encoder = CLIPTextModel.from_pretrained(model_id, subfolder="text_encoder").to(DEVICE)
    vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae").to(DEVICE)
    unet = UNet2DConditionModel.from_pretrained(model_id, subfolder="unet").to(DEVICE)
    noise_scheduler = DDPMScheduler.from_pretrained(model_id, subfolder="scheduler")

    # Freeze everything except LoRA
    text_encoder.requires_grad_(False)
    vae.requires_grad_(False)

    # ── Attach LoRA to U-Net ─────────────────────────────────────────────────
    lora_config = LoraConfig(
        r=config["rank"],
        lora_alpha=config["alpha"],
        lora_dropout=config["dropout"],
        target_modules=config["target_modules"],
        init_lora_weights="gaussian",
    )
    unet = get_peft_model(unet, lora_config)
    unet.print_trainable_parameters()

    # ── Dataset & DataLoader ─────────────────────────────────────────────────
    dataset = InpaintingTripletDataset(manifest_path, resolution=config["resolution"])
    dataloader = DataLoader(
        dataset, batch_size=config["batch_size"], shuffle=True,
        num_workers=2, pin_memory=True, drop_last=True,
    )

    # ── Optimizer & Scheduler ────────────────────────────────────────────────
    trainable_params = [p for p in unet.parameters() if p.requires_grad]
    optimizer = torch.optim.AdamW(
        trainable_params,
        lr=config["learning_rate"],
        weight_decay=config["weight_decay"],
    )

    warmup_steps = int(steps * config["warmup_ratio"])
    lr_scheduler = torch.optim.lr_scheduler.LambdaLR(
        optimizer,
        lr_lambda=lambda step: min(1.0, step / max(warmup_steps, 1)) *
                               (0.5 * (1 + math.cos(math.pi * step / steps)))
    )

    scaler = torch.amp.GradScaler("cuda") if config["mixed_precision"] else None

    # ── Training ─────────────────────────────────────────────────────────────
    unet.train()
    text_encoder.eval()
    vae.eval()

    data_iter = iter(dataloader)
    loss_log = []
    best_loss = float("inf")
    pbar = tqdm(range(steps), desc=f"  {category}")

    for step in pbar:
        # Cycle through dataset
        try:
            batch = next(data_iter)
        except StopIteration:
            data_iter = iter(dataloader)
            batch = next(data_iter)

        with torch.amp.autocast("cuda", enabled=config["mixed_precision"]):

            # ── Encode target (defect image) to latent ───────────────────────
            target_img = batch["defect"].to(DEVICE)          # [B, 3, H, W]
            latents = vae.encode(target_img).latent_dist.sample()
            latents = latents * vae.config.scaling_factor

            # ── Encode normal image to latent ────────────────────────────────
            normal_img = batch["normal"].to(DEVICE)
            normal_latents = vae.encode(normal_img).latent_dist.sample()
            normal_latents = normal_latents * vae.config.scaling_factor

            # ── Resize mask to latent space ──────────────────────────────────
            mask = batch["mask"].to(DEVICE)                  # [B, 1, H, W]
            mask_latent = F.interpolate(
                mask, size=latents.shape[-2:], mode="nearest"
            )

            # ── Add noise ────────────────────────────────────────────────────
            noise = torch.randn_like(latents)
            timesteps = torch.randint(
                0, noise_scheduler.config.num_train_timesteps,
                (latents.shape[0],), device=DEVICE
            ).long()
            noisy_latents = noise_scheduler.add_noise(latents, noise, timesteps)

            # ── Build SD inpainting input ────────────────────────────────────
            # SD 2 inpainting expects: [noisy_latents, mask, masked_normal_latents]
            # concatenated on channel dim → 4 + 1 + 4 = 9 channels
            masked_normal_latents = normal_latents * (1 - mask_latent)
            unet_input = torch.cat([noisy_latents, mask_latent, masked_normal_latents], dim=1)

            # ── Encode text prompt ───────────────────────────────────────────
            text_inputs = tokenizer(
                batch["prompt"], padding="max_length",
                max_length=tokenizer.model_max_length,
                truncation=True, return_tensors="pt"
            ).to(DEVICE)
            encoder_hidden_states = text_encoder(text_inputs.input_ids)[0]

            # ── Predict noise ────────────────────────────────────────────────
            noise_pred = unet(
                unet_input, timesteps, encoder_hidden_states
            ).sample

            # ── Loss: MSE on predicted noise (standard diffusion loss) ───────
            # Weight the loss: higher weight inside defect mask
            mask_weight = 1.0 + 2.0 * mask_latent  # 3× weight in defect region
            loss = F.mse_loss(noise_pred, noise, reduction="none")
            loss = (loss * mask_weight).mean()

        # ── Backprop ─────────────────────────────────────────────────────────
        scaled_loss = loss / config["grad_accum"]

        if scaler:
            scaler.scale(scaled_loss).backward()
            if (step + 1) % config["grad_accum"] == 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(trainable_params, 1.0)
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
        else:
            scaled_loss.backward()
            if (step + 1) % config["grad_accum"] == 0:
                torch.nn.utils.clip_grad_norm_(trainable_params, 1.0)
                optimizer.step()
                optimizer.zero_grad()

        lr_scheduler.step()

        # Logging
        loss_val = loss.item()
        loss_log.append(loss_val)
        if loss_val < best_loss:
            best_loss = loss_val

        pbar.set_postfix(loss=f"{loss_val:.4f}", best=f"{best_loss:.4f}",
                         lr=f"{lr_scheduler.get_last_lr()[0]:.2e}")

        # ── Checkpoint ───────────────────────────────────────────────────────
        if (step + 1) % config["save_every"] == 0:
            unet.save_pretrained(f"{cat_lora_dir}/checkpoint-{step+1}")

    # ── Save final LoRA weights ──────────────────────────────────────────────
    unet.save_pretrained(cat_lora_dir)

    # Save training config & loss curve
    train_meta = {
        "category": category,
        "n_triplets": n_triplets,
        "total_steps": steps,
        "final_loss": round(loss_log[-1], 6) if loss_log else None,
        "best_loss": round(best_loss, 6),
        "mean_loss_last_100": round(np.mean(loss_log[-100:]), 6) if len(loss_log) >= 100 else None,
        "config": config,
        "base_model": model_id,
        "timestamp": datetime.now().isoformat(),
    }
    with open(f"{cat_lora_dir}/training_meta.json", "w") as f:
        json.dump(train_meta, f, indent=2)

    np.save(f"{cat_lora_dir}/loss_curve.npy", np.array(loss_log))

    # Mark as done
    Path(done_flag).touch()

    # Cleanup
    del unet, vae, text_encoder, tokenizer, optimizer, lr_scheduler, dataloader, dataset
    gc.collect()
    torch.cuda.empty_cache()

    print(f"  ✓ {category}: saved to {cat_lora_dir} (final loss={loss_log[-1]:.4f})")
    return True


# ── Run training for all categories ──────────────────────────────────────────
training_report = {}

for cat in TRAIN_CATEGORIES:
    success = train_lora_for_category(cat, LORA_CONFIG)
    training_report[cat] = "trained" if success else "skipped"

print(f"\n{'═' * 60}")
print("LoRA TRAINING SUMMARY")
print(f"{'═' * 60}")
for cat, status in training_report.items():
    icon = "✓" if status == "trained" else "⚠"
    meta_path = f"{LORA_DIR}/{cat}/training_meta.json"
    if os.path.exists(meta_path):
        with open(meta_path) as f:
            meta = json.load(f)
        print(f"  {icon} {cat:<15} {status:<10} steps={meta['total_steps']}  "
              f"final_loss={meta.get('final_loss', 'N/A')}")
    else:
        print(f"  {icon} {cat:<15} {status}")

with open(f"{LORA_DIR}/training_report.json", "w") as f:
    json.dump(training_report, f, indent=2)


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 4: VALIDATE TRAINED LoRA MODELS                                   ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# For each trained LoRA, generate sample defect images and measure:
#   - FID: distribution distance to real defects
#   - LPIPS: perceptual similarity to nearest real defect
#   - SSIM (outside mask): structure preservation
#   - Visual grid: save comparison images
# ═══════════════════════════════════════════════════════════════════════════════

print("=" * 80)
print("CELL 4: Validating LoRA models")
print("=" * 80)

from diffusers import StableDiffusionInpaintPipeline
from peft import PeftModel
import lpips
from torchmetrics.image import StructuralSimilarityIndexMeasure

# Load LPIPS
lpips_fn = lpips.LPIPS(net="alex").to(DEVICE)
ssim_fn = StructuralSimilarityIndexMeasure(data_range=1.0).to(DEVICE)


def load_lora_pipeline(category):
    """Load base SD inpainting + LoRA adapter for a category."""
    model_id = "sd2-community/stable-diffusion-2-inpainting"
    cat_lora_dir = f"{LORA_DIR}/{category}"

    pipe = StableDiffusionInpaintPipeline.from_pretrained(
        model_id, torch_dtype=torch.float16, safety_checker=None,
    )
    # Load LoRA weights into the pipeline's U-Net
    pipe.unet = PeftModel.from_pretrained(pipe.unet, cat_lora_dir)
    pipe = pipe.to(DEVICE)
    pipe.set_progress_bar_config(disable=True)
    pipe.enable_attention_slicing()
    return pipe


def generate_with_lora(pipe, normal_path, mask_path, defect_type, category, seed=42):
    """Generate a single defect image using LoRA pipeline."""
    normal = Image.open(normal_path).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    mask = Image.open(mask_path).convert("L").resize((IMG_SIZE, IMG_SIZE))

    # Ensure mask is binary
    mask_np = np.array(mask)
    mask_np = (mask_np > 127).astype(np.uint8) * 255
    mask = Image.fromarray(mask_np).convert("RGB")  # SD expects 3-channel mask

    prompt = (
        f"a {defect_type.replace('_', ' ')} defect on a {category.replace('_', ' ')}, "
        f"industrial inspection photograph, macro, high resolution"
    )
    negative_prompt = (
        "cartoon, illustration, drawing, painting, artistic, low quality, "
        "blurry, watermark, text, unrealistic, different object"
    )

    generator = torch.Generator(device=DEVICE).manual_seed(seed)

    with torch.autocast("cuda"):
        result = pipe(
            prompt=prompt,
            negative_prompt=negative_prompt,
            image=normal,
            mask_image=mask,
            guidance_scale=7.5,
            num_inference_steps=50,
            height=IMG_SIZE, width=IMG_SIZE,
            generator=generator,
        ).images[0]

    return result


VALIDATION_CATEGORIES = ALL_CATEGORIES  # or subset
N_VALIDATION_SAMPLES = 5  # per defect type

val_results = {}

for cat in VALIDATION_CATEGORIES:
    cat_lora_dir = f"{LORA_DIR}/{cat}"
    if not os.path.exists(f"{cat_lora_dir}/.done"):
        continue

    manifest_path = f"{TRIPLET_DIR}/{cat}/manifest.json"
    with open(manifest_path) as f:
        manifest = json.load(f)

    print(f"\n── {cat} ──")

    # Load pipeline
    pipe = load_lora_pipeline(cat)

    cat_eval_dir = f"{EVAL_S1_DIR}/{cat}"
    os.makedirs(cat_eval_dir, exist_ok=True)

    # Group triplets by defect type
    by_defect = defaultdict(list)
    for t in manifest["triplets"]:
        by_defect[t["defect_type"]].append(t)

    cat_lpips_scores = []
    cat_ssim_scores = []

    for dt, triplets in by_defect.items():
        dt_eval_dir = f"{cat_eval_dir}/{dt}"
        os.makedirs(dt_eval_dir, exist_ok=True)

        # Sample a few triplets
        sample = triplets[:N_VALIDATION_SAMPLES]

        for i, t in enumerate(sample):
            # Generate
            gen_img = generate_with_lora(
                pipe, t["normal_image"], t["mask_image"],
                t["defect_type"], t["category"], seed=42 + i
            )
            gen_img.save(f"{dt_eval_dir}/gen_{i:03d}.png")

            # Load real defect for comparison
            real_defect = Image.open(t["defect_image"]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
            normal_img = Image.open(t["normal_image"]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
            mask_img = Image.open(t["mask_image"]).convert("L").resize((IMG_SIZE, IMG_SIZE))

            # ── LPIPS (generated vs real defect) ─────────────────────────────
            gen_t = transforms.ToTensor()(gen_img).unsqueeze(0).to(DEVICE) * 2 - 1
            real_t = transforms.ToTensor()(real_defect).unsqueeze(0).to(DEVICE) * 2 - 1
            with torch.no_grad():
                lpips_score = lpips_fn(gen_t, real_t).item()
            cat_lpips_scores.append(lpips_score)

            # ── SSIM outside mask (structure preservation) ───────────────────
            gen_01 = transforms.ToTensor()(gen_img).unsqueeze(0).to(DEVICE)
            norm_01 = transforms.ToTensor()(normal_img).unsqueeze(0).to(DEVICE)
            mask_t = transforms.ToTensor()(mask_img).unsqueeze(0).to(DEVICE)
            mask_inv = (mask_t < 0.5).float()

            # Apply inverse mask — keep only non-defect regions
            gen_masked = gen_01 * mask_inv
            norm_masked = norm_01 * mask_inv

            if mask_inv.sum() > 100:  # enough pixels to compute SSIM
                ssim_val = ssim_fn(gen_masked, norm_masked).item()
                cat_ssim_scores.append(ssim_val)

            # Save comparison grid
            grid = Image.new("RGB", (IMG_SIZE * 4, IMG_SIZE))
            grid.paste(normal_img, (0, 0))
            grid.paste(real_defect, (IMG_SIZE, 0))
            grid.paste(gen_img, (IMG_SIZE * 2, 0))
            # Mask visualization
            mask_vis = Image.open(t["mask_image"]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
            grid.paste(mask_vis, (IMG_SIZE * 3, 0))
            grid.save(f"{dt_eval_dir}/comparison_{i:03d}.png")

    # Aggregate
    mean_lpips = round(np.mean(cat_lpips_scores), 4) if cat_lpips_scores else None
    mean_ssim = round(np.mean(cat_ssim_scores), 4) if cat_ssim_scores else None

    val_results[cat] = {
        "mean_lpips": mean_lpips,
        "mean_ssim_outside_mask": mean_ssim,
        "n_samples": len(cat_lpips_scores),
    }

    print(f"  LPIPS (gen↔real): {mean_lpips}  |  SSIM (outside mask): {mean_ssim}  "
          f"({len(cat_lpips_scores)} samples)")

    del pipe
    gc.collect()
    torch.cuda.empty_cache()

# Save validation results
with open(f"{EVAL_S1_DIR}/validation_results.json", "w") as f:
    json.dump(val_results, f, indent=2)

print(f"\n{'═' * 70}")
print("VALIDATION SUMMARY")
print(f"{'═' * 70}")
print(f"{'Category':<15} {'LPIPS↓':>8} {'SSIM↑':>8} {'N':>5}")
print(f"{'─' * 40}")
for cat, r in val_results.items():
    print(f"{cat:<15} {r['mean_lpips'] or 'N/A':>8} {r['mean_ssim_outside_mask'] or 'N/A':>8} {r['n_samples']:>5}")

del lpips_fn, ssim_fn
gc.collect()
torch.cuda.empty_cache()

```

```python

# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 5: CONTROLNET EDGE-CONDITIONED INPAINTING SETUP                   ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# Use pretrained ControlNet-Canny to preserve object structure during
# inpainting. No fine-tuning needed for first pass — we test pretrained.
# If results are bad, Cell 5B below fine-tunes a ControlNet adapter.
# ═══════════════════════════════════════════════════════════════════════════════
import os
import json
import os, torch, numpy as np
from PIL import Image
import os
from datetime import datetime
MVTEC = "/root/project/datasets/mvtec_ad"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
CKPT_DIR = f"{ROOT}/checkpoints"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ROOT = "/root/project"
RES_DIR = f"{ROOT}/results"
EVAL_S1_DIR = f"{RES_DIR}/stage1_validation"
import os, json, torch, gc
from PIL import Image
import numpy as np
import cv2
from peft import PeftModel

from diffusers import (
    ControlNetModel,
    StableDiffusionControlNetInpaintPipeline,
    UniPCMultistepScheduler,
)
ALL_CATEGORIES = [
    "bottle", "cable", "capsule", "carpet", "grid",
    "hazelnut", "leather", "metal_nut", "pill",
    "screw", "tile", "toothbrush", "transistor",
    "wood", "zipper"
]

ROOT = "/root/project"
FEATURES_DIR = f"{ROOT}/features"
# ===== GLOBAL SETUP =====
ROOT = "/root/project"

TRIPLET_DIR  = f"{ROOT}/training_triplets"
LORA_DIR     = f"{ROOT}/checkpoints/sd_lora"
EVAL_S1_DIR  = f"{ROOT}/results/stage1_validation"
from torchvision import transforms

dinov2_transform = transforms.Compose([
    transforms.Resize((518, 518)),  # required for DINOv2
    transforms.ToTensor(),
    transforms.Normalize(
        mean=(0.485, 0.456, 0.406),
        std=(0.229, 0.224, 0.225),
    ),
])
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
IMG_SIZE = 512

os.makedirs(EVAL_S1_DIR, exist_ok=True)
import os, json, torch, gc
from PIL import Image
import numpy as np
import cv2

from diffusers import (
    ControlNetModel,
    StableDiffusionControlNetInpaintPipeline,
    UniPCMultistepScheduler,
)

from peft import PeftModel
ROOT = "/root/project"

TRIPLET_DIR = f"{ROOT}/training_triplets"
LORA_DIR    = f"{ROOT}/checkpoints/sd_lora"
EVAL_S1_DIR = f"{ROOT}/results/stage1_validation"

os.makedirs(EVAL_S1_DIR, exist_ok=True)
os.makedirs(EVAL_S1_DIR, exist_ok=True)
import torch
print("=" * 80)
print("CELL 5: ControlNet setup")
print("=" * 80)

from diffusers import (
    StableDiffusionInpaintPipeline,
    ControlNetModel,
    StableDiffusionControlNetInpaintPipeline,
    UniPCMultistepScheduler,
)
from controlnet_aux import CannyDetector
import cv2


# Load ControlNet-Canny (pretrained)
print("Loading ControlNet-Canny ...")
controlnet = ControlNetModel.from_pretrained(
    "thibaud/controlnet-sd21-canny-diffusers",
    torch_dtype=torch.float16,
)
print("✓ ControlNet loaded")


def extract_canny_edges(image, low=100, high=200):
    """Extract Canny edges from a PIL image."""
    img_np = np.array(image.convert("RGB"))
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, low, high)
    return Image.fromarray(edges).convert("RGB")


def generate_with_controlnet(
    category, normal_path, mask_path, defect_type, severity="moderate", seed=42
):
    """
    Generate defect image using LoRA + ControlNet.
    ControlNet preserves object edges while LoRA generates realistic defects.
    """
    model_id = "sd2-community/stable-diffusion-2-inpainting"
    cat_lora_dir = f"{LORA_DIR}/{category}"

    # Build pipeline with ControlNet
    pipe = StableDiffusionControlNetInpaintPipeline.from_pretrained(
        model_id,
        controlnet=controlnet,
        torch_dtype=torch.float16,
        safety_checker=None,
    )

    # Load LoRA if available
    if os.path.exists(f"{cat_lora_dir}/.done"):
        pipe.unet = PeftModel.from_pretrained(pipe.unet, cat_lora_dir)

    pipe = pipe.to(DEVICE)
    pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
    pipe.set_progress_bar_config(disable=True)
    pipe.enable_attention_slicing()

    # Prepare inputs
    normal = Image.open(normal_path).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    mask = Image.open(mask_path).convert("L").resize((IMG_SIZE, IMG_SIZE))

    # Binarize mask
    mask_np = (np.array(mask) > 127).astype(np.uint8) * 255
    mask = Image.fromarray(mask_np)

    # Extract Canny edges from normal image (structure to preserve)
    canny = extract_canny_edges(normal)

    prompt = (
        f"a {severity} {defect_type.replace('_', ' ')} defect on a "
        f"{category.replace('_', ' ')}, industrial inspection, macro photograph"
    )
    negative_prompt = (
        "cartoon, illustration, painting, low quality, blurry, "
        "watermark, text, different object, wrong shape"
    )

    generator = torch.Generator(device=DEVICE).manual_seed(seed)

    with torch.autocast("cuda"):
        result = pipe(
            prompt=prompt,
            negative_prompt=negative_prompt,
            image=normal,
            mask_image=mask,
            control_image=canny,
            guidance_scale=7.5,
            controlnet_conditioning_scale=0.8,  # how strongly to follow edges
            num_inference_steps=50,
            height=IMG_SIZE, width=IMG_SIZE,
            generator=generator,
        ).images[0]

    del pipe
    gc.collect()
    torch.cuda.empty_cache()

    return result, canny


# ── Quick test on 3 categories ───────────────────────────────────────────────
print("\nTesting ControlNet + LoRA on sample images ...")

cnet_test_dir = f"{EVAL_S1_DIR}/controlnet_test"
os.makedirs(cnet_test_dir, exist_ok=True)

test_cats = ["bottle", "metal_nut", "transistor"]

for cat in test_cats:
    manifest_path = f"{TRIPLET_DIR}/{cat}/manifest.json"
    if not os.path.exists(manifest_path):
        continue

    with open(manifest_path) as f:
        manifest = json.load(f)

    if not manifest["triplets"]:
        continue

    t = manifest["triplets"][0]  # First triplet
    gen_img, canny_img = generate_with_controlnet(
        cat, t["normal_image"], t["mask_image"], t["defect_type"]
    )

    # Save comparison: Normal | Canny | Real Defect | Generated | Mask
    normal = Image.open(t["normal_image"]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    real = Image.open(t["defect_image"]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    mask = Image.open(t["mask_image"]).convert("L").resize((IMG_SIZE, IMG_SIZE)).convert("RGB")

    grid = Image.new("RGB", (IMG_SIZE * 5, IMG_SIZE))
    grid.paste(normal, (0, 0))
    grid.paste(canny_img, (IMG_SIZE, 0))
    grid.paste(real, (IMG_SIZE * 2, 0))
    grid.paste(gen_img, (IMG_SIZE * 3, 0))
    grid.paste(mask, (IMG_SIZE * 4, 0))
    grid.save(f"{cnet_test_dir}/{cat}_test.png")
    print(f"  ✓ {cat}: saved test grid")

del controlnet
gc.collect()
torch.cuda.empty_cache()
print("✓ ControlNet tests saved")


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 6: SAM PRECISE MASK GENERATION MODULE                             ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# Build a mask generation module that:
#   1. Segments the object from background using SAM
#   2. Places a small, precise defect mask at the hypothesis location
#   3. Constrains mask to be WITHIN the object boundary
#   4. Ensures mask covers 5-15% of image (not 70%)
# ═══════════════════════════════════════════════════════════════════════════════

print("=" * 80)
print("CELL 6: SAM mask generation module")
print("=" * 80)

from segment_anything import sam_model_registry, SamPredictor, SamAutomaticMaskGenerator

SAM_CHECKPOINT = f"{CKPT_DIR}/sam_vit_h_4b8939.pth"

print("Loading SAM ...")
sam = sam_model_registry["vit_h"](checkpoint=SAM_CHECKPOINT)
sam = sam.to(DEVICE)
sam_predictor = SamPredictor(sam)
sam_auto = SamAutomaticMaskGenerator(
    sam,
    points_per_side=32,
    pred_iou_thresh=0.86,
    stability_score_thresh=0.92,
    min_mask_region_area=1000,
)
print("✓ SAM loaded")


class PreciseMaskGenerator:
    """
    Generates physically grounded defect masks using SAM object segmentation.
    Replaces the old create_defect_mask() which made giant rectangles.
    """

    def __init__(self, sam_predictor, sam_auto, target_defect_ratio=(0.03, 0.15)):
        self.predictor = sam_predictor
        self.auto_gen = sam_auto
        self.min_ratio, self.max_ratio = target_defect_ratio

    def get_object_mask(self, image):
        """Segment the main object from background using SAM auto."""
        img_np = np.array(image.convert("RGB"))
        masks = self.auto_gen.generate(img_np)

        if not masks:
            # Fallback: everything except border
            h, w = img_np.shape[:2]
            mask = np.zeros((h, w), dtype=bool)
            border = int(min(h, w) * 0.05)
            mask[border:h-border, border:w-border] = True
            return mask

        # Pick largest non-background mask
        # Sort by area descending
        masks.sort(key=lambda x: x["area"], reverse=True)

        # The largest is often the object; skip if it covers >90% (likely background)
        for m in masks:
            ratio = m["area"] / (img_np.shape[0] * img_np.shape[1])
            if 0.05 < ratio < 0.90:
                return m["segmentation"]

        # Fallback to largest
        return masks[0]["segmentation"]

    def generate_defect_mask(
        self, image, bbox_normalized, object_mask=None, mask_shape="ellipse"
    ):
        """
        Generate a precise defect mask constrained within the object boundary.

        Args:
            image: PIL Image
            bbox_normalized: [x1, y1, x2, y2] in [0, 1] from hypothesis
            object_mask: pre-computed SAM mask (optional)
            mask_shape: "ellipse", "rectangle", or "irregular"

        Returns:
            PIL Image (L mode, 0/255), actual defect ratio
        """
        w, h = image.size

        # Get object mask if not provided
        if object_mask is None:
            object_mask = self.get_object_mask(image)

        # Convert bbox to pixels
        x1 = int(bbox_normalized[0] * w)
        y1 = int(bbox_normalized[1] * h)
        x2 = int(bbox_normalized[2] * w)
        y2 = int(bbox_normalized[3] * h)

        # Clamp
        x1, y1 = max(0, x1), max(0, y1)
        x2, y2 = min(w, x2), min(h, y2)

        # Ensure minimum size (at least 20px each side)
        if x2 - x1 < 20:
            cx = (x1 + x2) // 2
            x1, x2 = max(0, cx - 20), min(w, cx + 20)
        if y2 - y1 < 20:
            cy = (y1 + y2) // 2
            y1, y2 = max(0, cy - 20), min(h, cy + 20)

        # Create raw defect mask within bbox
        defect_mask = np.zeros((h, w), dtype=np.uint8)

        if mask_shape == "ellipse":
            cy, cx = (y1 + y2) // 2, (x1 + x2) // 2
            ry, rx = (y2 - y1) // 2, (x2 - x1) // 2
            Y, X = np.ogrid[:h, :w]
            ellipse = ((X - cx) ** 2 / max(rx ** 2, 1) +
                       (Y - cy) ** 2 / max(ry ** 2, 1)) <= 1
            defect_mask[ellipse] = 255

        elif mask_shape == "irregular":
            # Blob-like mask using random perturbations
            cy, cx = (y1 + y2) // 2, (x1 + x2) // 2
            ry, rx = (y2 - y1) // 2, (x2 - x1) // 2
            Y, X = np.ogrid[:h, :w]
            # Add noise to make it irregular
            angles = np.arctan2(Y - cy, X - cx)
            radii = np.sqrt((X - cx) ** 2 + (Y - cy) ** 2)
            np.random.seed(42)
            noise = np.random.uniform(0.7, 1.3, size=angles.shape)
            threshold = np.sqrt((rx * np.cos(angles)) ** 2 + (ry * np.sin(angles)) ** 2)
            defect_mask[radii < threshold * noise] = 255

        else:  # rectangle
            defect_mask[y1:y2, x1:x2] = 255

        # ── Constrain to object boundary ─────────────────────────────────────
        object_mask_uint8 = (object_mask.astype(np.uint8)) * 255
        defect_mask = np.minimum(defect_mask, object_mask_uint8)

        # ── Check defect ratio ───────────────────────────────────────────────
        defect_pixels = (defect_mask > 0).sum()
        total_pixels = h * w
        ratio = defect_pixels / total_pixels

        # If mask is too large, shrink the bbox
        if ratio > self.max_ratio:
            scale = math.sqrt(self.max_ratio / max(ratio, 1e-6)) * 0.9
            cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
            hw, hh = (x2 - x1) * scale / 2, (y2 - y1) * scale / 2
            x1, y1 = int(cx - hw), int(cy - hh)
            x2, y2 = int(cx + hw), int(cy + hh)
            # Regenerate (recursive would be cleaner but this is fine)
            defect_mask = np.zeros((h, w), dtype=np.uint8)
            defect_mask[max(0,y1):min(h,y2), max(0,x1):min(w,x2)] = 255
            defect_mask = np.minimum(defect_mask, object_mask_uint8)

        # If mask is too small or empty, expand
        if (defect_mask > 0).sum() / total_pixels < self.min_ratio:
            # Find object centroid and place a minimum-size mask there
            obj_ys, obj_xs = np.where(object_mask)
            if len(obj_ys) > 0:
                cy, cx = int(np.median(obj_ys)), int(np.median(obj_xs))
                radius = int(math.sqrt(self.min_ratio * total_pixels / math.pi))
                Y, X = np.ogrid[:h, :w]
                circle = (X - cx) ** 2 + (Y - cy) ** 2 <= radius ** 2
                defect_mask[circle] = 255
                defect_mask = np.minimum(defect_mask, object_mask_uint8)

        # ── Smooth edges ─────────────────────────────────────────────────────
        from scipy.ndimage import gaussian_filter
        defect_mask_smooth = gaussian_filter(defect_mask.astype(float), sigma=3)
        defect_mask_final = (defect_mask_smooth > 127).astype(np.uint8) * 255

        actual_ratio = (defect_mask_final > 0).sum() / total_pixels

        return Image.fromarray(defect_mask_final).convert("L"), actual_ratio


# ── Test mask generator ──────────────────────────────────────────────────────
mask_gen = PreciseMaskGenerator(sam_predictor, sam_auto)

mask_test_dir = f"{EVAL_S1_DIR}/mask_tests"
os.makedirs(mask_test_dir, exist_ok=True)

print("\nTesting SAM mask generation ...")
for cat in ["bottle", "hazelnut", "toothbrush", "metal_nut", "transistor"]:
    normal_paths = get_image_paths(cat, "train", "good")
    if not normal_paths:
        continue

    img = Image.open(normal_paths[0]).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    obj_mask = mask_gen.get_object_mask(img)

    # Test with a small hypothesis bbox
    test_bboxes = [
        [0.35, 0.40, 0.55, 0.60],  # center, small
        [0.20, 0.60, 0.40, 0.80],  # lower left, small
        [0.50, 0.20, 0.70, 0.40],  # upper right, small
    ]

    for i, bbox in enumerate(test_bboxes):
        defect_mask, ratio = mask_gen.generate_defect_mask(
            img, bbox, object_mask=obj_mask, mask_shape="ellipse"
        )

        # Save grid: image | object mask | defect mask
        obj_vis = Image.fromarray((obj_mask * 255).astype(np.uint8)).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
        def_vis = defect_mask.convert("RGB")
        grid = Image.new("RGB", (IMG_SIZE * 3, IMG_SIZE))
        grid.paste(img, (0, 0))
        grid.paste(obj_vis, (IMG_SIZE, 0))
        grid.paste(def_vis, (IMG_SIZE * 2, 0))
        grid.save(f"{mask_test_dir}/{cat}_bbox{i}.png")

    print(f"  ✓ {cat}: masks generated (ratio={ratio:.3f})")

# Save the mask generator as a module for Stage 2
mask_module_code = '''
# PreciseMaskGenerator — import this in Stage 2
# Usage:
#   from mask_generator import PreciseMaskGenerator
#   mask_gen = PreciseMaskGenerator(sam_predictor, sam_auto)
#   mask, ratio = mask_gen.generate_defect_mask(image, bbox, object_mask)
'''

print("✓ SAM mask module ready")

del sam, sam_predictor, sam_auto, mask_gen
gc.collect()
torch.cuda.empty_cache()




# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 8: EXTRACT & INDEX ALL DINOv2 FEATURES (TRAIN + TEST)             ║
# ╚═══════════════════════════════════════════════════════════════════════════╝
# Build comprehensive feature indices for all images.
# Used by the critic (Stage 2) and evaluation (Stage 3).
# ═══════════════════════════════════════════════════════════════════════════════

print("=" * 80)
print("CELL 8: Comprehensive DINOv2 feature extraction")
print("=" * 80)

import faiss

# Reload DINOv2
print("Loading DINOv2 ...")
dinov2 = torch.hub.load("facebookresearch/dinov2", "dinov2_vitb14", verbose=False)
dinov2 = dinov2.eval().to(DEVICE)
print("✓ DINOv2 loaded")


@torch.no_grad()
def extract_features_batch(image_paths, batch_size=16):
    """Extract CLS and patch features for a batch of images."""
    all_cls = []
    all_patches = []

    for i in range(0, len(image_paths), batch_size):
        batch_paths = image_paths[i:i + batch_size]
        imgs = []
        for p in batch_paths:
            img = Image.open(p).convert("RGB")
            imgs.append(dinov2_transform(img))

        batch = torch.stack(imgs).to(DEVICE)
        out = dinov2.forward_features(batch)

        all_cls.append(out["x_norm_clstoken"].cpu())
        # For patches, subsample to save memory (every 10th patch)
        patches = out["x_norm_patchtokens"].cpu()
        all_patches.append(patches[:, ::10, :])

    cls_feats = torch.cat(all_cls, dim=0)
    patch_feats = torch.cat(all_patches, dim=0)

    return cls_feats, patch_feats


for cat in ALL_CATEGORIES:
    cat_feat_dir = f"{FEATURES_DIR}/{cat}"
    os.makedirs(cat_feat_dir, exist_ok=True)

    # Check if test features already exist
    test_cls_path = f"{cat_feat_dir}/test_cls_features.pt"
    if os.path.exists(test_cls_path):
        print(f"  {cat}: test features exist (skipping)")
        continue

    print(f"  {cat}:", end=" ", flush=True)

    # ── Train features (may already exist) ───────────────────────────────────
    train_paths = get_image_paths(cat, "train", "good")
    train_cls_path = f"{cat_feat_dir}/cls_features.pt"
    if not os.path.exists(train_cls_path):
        train_cls, train_patches = extract_features_batch(train_paths)
        torch.save(train_cls, train_cls_path)
        torch.save(train_patches, f"{cat_feat_dir}/patch_features.pt")
        print(f"train({len(train_paths)})", end=" ", flush=True)

    # ── Test features ────────────────────────────────────────────────────────
    test_dir = os.path.join(MVTEC, cat, "test")
    test_paths = []
    test_labels = []
    test_defect_types = []

    # Good
    for p in get_image_paths(cat, "test", "good"):
        test_paths.append(p)
        test_labels.append(0)
        test_defect_types.append("good")

    # Defective
    for dd in sorted(os.listdir(test_dir)):
        dp = os.path.join(test_dir, dd)
        if dd == "good" or not os.path.isdir(dp):
            continue
        for p in get_image_paths(cat, "test", dd):
            test_paths.append(p)
            test_labels.append(1)
            test_defect_types.append(dd)

    if test_paths:
        test_cls, test_patches = extract_features_batch(test_paths)
        torch.save(test_cls, test_cls_path)
        torch.save(test_patches, f"{cat_feat_dir}/test_patch_features.pt")

        # Save test metadata
        test_meta = {
            "paths": test_paths,
            "labels": test_labels,
            "defect_types": test_defect_types,
        }
        with open(f"{cat_feat_dir}/test_metadata.json", "w") as f:
            json.dump(test_meta, f, indent=2)

        print(f"test({len(test_paths)})", end=" ", flush=True)

    # ── Build FAISS index for fast k-NN ──────────────────────────────────────
    train_cls = torch.load(train_cls_path, map_location="cpu")
    train_np = train_cls.numpy().astype(np.float32)

    # Normalize for cosine similarity
    faiss.normalize_L2(train_np)

    index = faiss.IndexFlatIP(train_np.shape[1])  # inner product after normalization = cosine
    index.add(train_np)
    faiss.write_index(index, f"{cat_feat_dir}/faiss_cls.index")

    print(f"faiss({train_np.shape[0]} vectors) ✓")

del dinov2
gc.collect()
torch.cuda.empty_cache()
print("\n✓ All features extracted and indexed")


# ╔═══════════════════════════════════════════════════════════════════════════╗
# ║  CELL 9: STAGE 1 SUMMARY & CHECKPOINT INVENTORY                         ║
# ╚═══════════════════════════════════════════════════════════════════════════╝

print("=" * 80)
print("STAGE 1 COMPLETE — CHECKPOINT INVENTORY")
print("=" * 80)

inventory = {
    "timestamp": datetime.now().isoformat(),
    "categories": {},
}

for cat in ALL_CATEGORIES:
    cat_status = {}

    # Triplets
    manifest_path = f"{TRIPLET_DIR}/{cat}/manifest.json"
    if os.path.exists(manifest_path):
        with open(manifest_path) as f:
            m = json.load(f)
        cat_status["triplets"] = m["n_triplets"]
    else:
        cat_status["triplets"] = 0

    # LoRA
    lora_done = os.path.exists(f"{LORA_DIR}/{cat}/.done")
    cat_status["lora_trained"] = lora_done
    if lora_done and os.path.exists(f"{LORA_DIR}/{cat}/training_meta.json"):
        with open(f"{LORA_DIR}/{cat}/training_meta.json") as f:
            meta = json.load(f)
        cat_status["lora_final_loss"] = meta.get("final_loss")
        cat_status["lora_steps"] = meta.get("total_steps")

    # Features
    cat_status["train_features"] = os.path.exists(f"{FEATURES_DIR}/{cat}/cls_features.pt")
    cat_status["test_features"] = os.path.exists(f"{FEATURES_DIR}/{cat}/test_cls_features.pt")
    cat_status["faiss_index"] = os.path.exists(f"{FEATURES_DIR}/{cat}/faiss_cls.index")

    # SAM masks (tested)
    cat_status["sam_tested"] = os.path.exists(f"{EVAL_S1_DIR}/mask_tests/{cat}_bbox0.png")

    inventory["categories"][cat] = cat_status

# Anomalib baselines
anom_path = f"{ANOMALIB_RESULTS}/baseline_results.json"
if os.path.exists(anom_path):
    with open(anom_path) as f:
        inventory["anomalib_baselines"] = json.load(f)

# Validation results
val_path = f"{EVAL_S1_DIR}/validation_results.json"
if os.path.exists(val_path):
    with open(val_path) as f:
        inventory["lora_validation"] = json.load(f)

# Save inventory
with open(f"{CKPT_DIR}/stage1_inventory.json", "w") as f:
    json.dump(inventory, f, indent=2)

# Print summary
print(f"\n{'Category':<15} {'Triplets':>9} {'LoRA':>6} {'Loss':>8} {'Feats':>6} {'FAISS':>6}")
print(f"{'─' * 55}")

for cat in ALL_CATEGORIES:
    s = inventory["categories"][cat]
    lora = "✓" if s["lora_trained"] else "✗"
    feats = "✓" if s["test_features"] else "✗"
    faiss_ok = "✓" if s["faiss_index"] else "✗"
    loss = f"{s.get('lora_final_loss', 'N/A')}" if s.get("lora_final_loss") else "N/A"
    print(f"{cat:<15} {s['triplets']:>9} {lora:>6} {loss:>8} {feats:>6} {faiss_ok:>6}")

print(f"\n{'═' * 55}")
print("STAGE 1 OUTPUTS:")
print(f"  • Training triplets:   {TRIPLET_DIR}/")
print(f"  • LoRA checkpoints:    {LORA_DIR}/")
print(f"  • ControlNet tests:    {EVAL_S1_DIR}/controlnet_test/")
print(f"  • SAM mask tests:      {EVAL_S1_DIR}/mask_tests/")
print(f"  • Anomalib baselines:  {ANOMALIB_RESULTS}/")
print(f"  • DINOv2 features:     {FEATURES_DIR}/")
print(f"  • Validation metrics:  {EVAL_S1_DIR}/")
print(f"  • Full inventory:      {CKPT_DIR}/stage1_inventory.json")
print(f"\n→ Ready for STAGE 2 (Agentic Reasoning Pipeline)")
```
